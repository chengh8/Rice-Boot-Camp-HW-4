```python
%matplotlib inline
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# File to Load (Remember to change these)
city_load = "city_data.csv"
ride_load = "data/ride_data.csv"

# Read the City and Ride Data
city_df = pd.read_csv(city_load)
ride_df = pd.read_csv(ride_load)

# Combine the data into a single dataset
city_ride_df = pd.merge(city_df, ride_df, on="city",how="outer")

# Display the data table for preview
city_ride_df.head()
```

```python
# Obtain the x and y coordinates for each of the three city types
Grp_type = city_ride_df.groupby(["type","city"])
Grp_type_unique = city_ride_df.drop_duplicates("city", keep='first')
Unique_grp = Grp_type_unique.groupby(["type","city"])

# created new dataframe with the desired data
New_Data = pd.DataFrame()
New_Data["Avg_fare"] = Grp_type["fare"].mean()
New_Data["Total_rides"] = Grp_type["ride_id"].count()
New_Data["Driver_count"] = Unique_grp["driver_count"].unique().apply(lambda x: x[0])

# split the data into group by index value (Rural, Urban, Suburban)
Rural_df = New_Data.iloc[New_Data.index.get_level_values('type') == "Rural"]
Urban_df = New_Data.iloc[New_Data.index.get_level_values('type') == "Urban"]
Suburban_df = New_Data.iloc[New_Data.index.get_level_values('type') == "Suburban"]

#Get the values in list form for the scatter plot
Rural_y = Rural_df["Avg_fare"].tolist()
Rural_x = Rural_df["Total_rides"].tolist()
Rural_size = Rural_df["Driver_count"].tolist()

#scale the scatter plot marks
Rural_s = [i * 10 for i in Rural_size]

Urban_y = Urban_df["Avg_fare"].tolist()
Urban_x = Urban_df["Total_rides"].tolist()
Urban_size = Urban_df["Driver_count"].tolist()

Urban_s = [i * 10 for i in Urban_size]

Suburban_y = Suburban_df["Avg_fare"].tolist()
Suburban_x = Suburban_df["Total_rides"].tolist()
Suburban_size = Suburban_df["Driver_count"].tolist()

Suburban_s = [i * 10 for i in Suburban_size]

# Build the scatter plots for each city types
Rural = plt.scatter(x=Rural_x, y=Rural_y, marker="o", facecolors=["yellow"], s = Rural_s, edgecolors="black", alpha=.4)
Urban = plt.scatter(x=Urban_x, y=Urban_y, marker="o", facecolors=["red"], s = Urban_s, edgecolors="black", alpha=.4)
Suburban = plt.scatter(x=Suburban_x, y=Suburban_y, marker="o", facecolors=["blue"], s = Suburban_s, edgecolors="black", alpha=.4)

# Incorporate the other graph properties
plt.title("Pyber Ride Sharing Data (2018)")
plt.xlabel("Total Number of Rides (per City)")
plt.ylabel("Average fare ($)")
plt.grid()

# Create a legend
lgnd = plt.legend((Rural,Urban,Suburban),("Rural","Urban","Suburban"),scatterpoints=1,loc='upper right', fontsize=10)

for handle in lgnd.legendHandles:
    handle.set_sizes([30.0])

# Incorporate a text label regarding circle size
plt.text(42, 40, "Note:")
plt.text(42, 38, "Circle size coorelates with driver count per city")

# Save Figure
plt.savefig("Pyber Ride Sharing Data (2018)")
plt.show()
```

For total number of rides and the number of drivers, the average fare is lower than cities with lower number of drivers and lower number of rides.

## Total Fares by City Type

```python
# Labels for the sections of pie chart
labels = ["Rural", "Urban", "Suburban"]

# The values of each section of the pie chart
Fare_Data = pd.DataFrame()
Fare_Data["Sum_fare"] = Grp_type["fare"].sum()

Total_Fares = Fare_Data["Sum_fare"].sum()

Rural_F = Fare_Data.iloc[Fare_Data.index.get_level_values('type') == "Rural"]
Urban_F = Fare_Data.iloc[Fare_Data.index.get_level_values('type') == "Urban"]
Suburban_F = Fare_Data.iloc[Fare_Data.index.get_level_values('type') == "Suburban"]

Rural_Fare_Total = Rural_F["Sum_fare"].sum()
Urban_Fare_Total = Urban_F["Sum_fare"].sum()
Suburban_Fare_Total = Suburban_F["Sum_fare"].sum()

Sizes = [Rural_Fare_Total, Urban_Fare_Total, Suburban_Fare_Total]

# The colors of each section of the pie chart
Colors = ["lightyellow", "lightcoral", "lightblue"]

# Seperate the Urban section from the others
Explode = [0, 0.1, 0]

plt.pie(Sizes, explode=Explode, labels=labels, colors=Colors, autopct="%1.1f%%", shadow=True, startangle=270)
plt.title("% of Total Fares by City Type")

plt.savefig("% of Total Fares by City Type")
plt.show()
```
